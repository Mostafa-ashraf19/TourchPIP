```python
import numpy as np
import h5py
from PIL import Image
import matplotlib.pyplot as plt
```

```python
CiFAR  = CIFAR_10(path='CIFAR_Data',download=False,train=True)
```

```python
dLoader = DataLoader(CiFAR,batchsize=500,shuffling=True,normalization={'Transform':False})
```

```python
image, labels = next(dLoader)
```

```python
def plot_CIFAR_normal(ind,images):
    arr = images[ind]
    sc_dpi = 157.35
    R = arr[0:1024].reshape(32,32)/255.0
    G = arr[1024:2048].reshape(32,32)/255.0
    B = arr[2048:].reshape(32,32)/255.0

    img = np.dstack((R,G,B))
    fig = plt.figure(figsize=(3,3))
    ax = fig.add_subplot(111)
    ax.imshow(img,interpolation='bicubic')
```

```python
def plot_CIFAR_afreshape(ind,images):
    arr = images[:,ind]
    sc_dpi = 157.35
    R = arr[0:1024].reshape(32,32)/255.0
    G = arr[1024:2048].reshape(32,32)/255.0
    B = arr[2048:].reshape(32,32)/255.0

    img = np.dstack((R,G,B))
    fig = plt.figure(figsize=(3,3))
    ax = fig.add_subplot(111)
    ax.imshow(img,interpolation='bicubic')
```

```python
plot_CIFAR_normal(20,image)
```
